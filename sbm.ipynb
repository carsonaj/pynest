```python
import numpy as np    
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, text
import networkx as nx

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
import pynest.rkg as rkg
import pynest.sbm as sbm
```

```python
lab_mat = np.zeros((1000, 2))
lab_mat[0:500, 0] = 1
lab_mat[500:1000, 1] = 1
```

```python
for p in [.51, .6, .7, .8, .9, .99]:
    ppm = sbm.PPM(p, lab_mat)
    a = rkg.samp_adj_mat(ppm.prob_mat)
    x_hat = rkg.get_adj_spec_emb(a, 2)
    plt.subplots(figsize=(6, 6))
    plt.scatter(x_hat[:, 0], x_hat[:, 1], s=1)
```
